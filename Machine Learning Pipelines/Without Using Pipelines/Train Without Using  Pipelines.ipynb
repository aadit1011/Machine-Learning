```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier


```

```python
df=df.drop(columns=['PassengerId','Cabin','Name','Ticket'])
df 
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2,random_state=42)
```

```python
si_age=SimpleImputer()
si_embarked=SimpleImputer(strategy='most_frequent')
x_train_age=si_age.fit_transform(x_train[['Age']])
x_train_embarked=si_embarked.fit_transform(x_train[['Embarked']])

x_test_age=si_age.transform(x_test[['Age']])
x_test_embarked=si_embarked.transform(x_test[['Embarked']])
```

```python
x_train_age
```

```python
x_train_embarked
```

```python
ohe_sex=OneHotEncoder(handle_unknown='ignore',sparse_output=False)
ohe_embarked=OneHotEncoder(handle_unknown='ignore',sparse_output=False)
```

```python
x_train_sex=ohe_sex.fit_transform(x_train[['Sex']])
x_train_embarked=ohe_embarked.fit_transform(x_train_embarked)

x_test_sex=ohe_sex.transform(x_test[['Sex']]) 
x_test_embarked=ohe_embarked.transform(x_test_embarked)
```

```python
x_train_embarked
```

```python
x_train_rem=x_train.drop(columns=['Age','Sex','Embarked'])
x_test_rem=x_test.drop(columns=['Age','Sex','Embarked'])
```

```python
x_train_rem
```

```python
x_test_rem
```

```python
x_train_final=np.concatenate((x_train_rem,x_train_age,x_train_sex,x_train_embarked),axis=1)
x_test_final=np.concatenate((x_test_rem,x_test_age,x_test_sex,x_test_embarked),axis=1)
```

```python
x_train_final
```

```python
x_test_final 
```

```python
x_test_final.shape 
```

```python
clf=DecisionTreeClassifier()
clf.fit(x_train_final,y_train)
```

```python
y_pred=clf.predict(x_test_final)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
