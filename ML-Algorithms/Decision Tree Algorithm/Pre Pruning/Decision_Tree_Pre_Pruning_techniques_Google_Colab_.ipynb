# Training the Model

```python
clf=DecisionTreeClassifier()
```

```python
parameter={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[3,4,5,6,7],
    'max_features':['auto','sqrt','log2'],
    'ccp_alpha': [0.0, 0.001, 0.01, 0.1]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf=DecisionTreeClassifier()
cv=GridSearchCV(clf,param_grid=parameter,scoring='accuracy',cv=6,n_jobs=-1)
```

```python
cv.fit(x_train,y_train)
```

```python
import pickle

# Save the entire GridSearchCV object
with open('/content/data/gridsearch_cv.pkl', 'wb') as f:
    pickle.dump(cv, f)

# Save just the best parameters
best_params = cv.best_params_
with open('/content/data/best_params.pkl', 'wb') as f:
    pickle.dump(best_params, f)

# Save the best estimator
best_estimator = cv.best_estimator_
with open('/content/data/best_decision_tree.pkl', 'wb') as f:
    pickle.dump(best_estimator, f)
```
