```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV


california=fetch_california_housing()

x,y=fetch_california_housing(return_X_y=True)

california.target_names

df=pd.DataFrame(california.data,columns=california.feature_names)


x=df.iloc[:,:]
y=pd.DataFrame(california.target,columns=california.target_names)

x

y



# # Applying Train test Split



x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)



# # Training The MOdel WIthout Hyper Parameter Tuning



dr=DecisionTreeRegressor(random_state=32)

dr.fit(x_train,y_train)




```

```python
y_pred=dr.predict(x_test)
print('R2 Score=',r2_score(y_test,y_pred))

```

```python
# Training Model WIth HyperParameter Tuning

model=DecisionTreeRegressor()

parameters={
    'criterion': ['squared_error', 'absolute_error'],
    'splitter': ['best'],
    'max_depth': [None, 6, 8, 10],
    'min_samples_split': [20, 30, 50],
    'min_samples_leaf': [10, 20, 30],
    'max_features': ['sqrt', 0.5, 0.7],
    'ccp_alpha': [0.01, 0.001, 0.0001]
}
```

```python
cv=GridSearchCV(model,param_grid=parameters,n_jobs=-1,cv=5,scoring='r2',return_train_score=True)
```

```python
cv.fit(x_train,y_train)
```

```python
import pickle as pkl
from google.colab import files

# Save parameters
with open('/content/data/parameters.pkl', 'wb') as f:
    pkl.dump(cv, f)

# Save best parameters
with open('/content/data/best_parameters.pkl', 'wb') as f:
    pkl.dump(cv.best_params_, f)

# Save best score
with open('/content/data/best_score.pkl', 'wb') as f:
    pkl.dump(cv.best_score_, f)

# Save best estimator
with open('/content/data/best_estimator.pkl', 'wb') as f:
    pkl.dump(cv.best_estimator_, f)

# Save best index
with open('/content/data/best_index.pkl', 'wb') as f:
    pkl.dump(cv.best_index_, f)

# Download the files
files.download('/content/data/parameters.pkl')
files.download('/content/data/best_parameters.pkl')
files.download('/content/data/best_score.pkl')
files.download('/content/data/best_estimator.pkl')
files.download('/content/data/best_index.pkl')
```
