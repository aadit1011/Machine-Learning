```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
from scipy import stats
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV

```

```python
california=fetch_california_housing()
```

```python
x,y=fetch_california_housing(return_X_y=True)
```

```python
california.target_names
```

```python
df=pd.DataFrame(california.data,columns=california.feature_names)
```

```python
df
```

```python
x=df.iloc[:,:]
y=pd.DataFrame(california.target,columns=california.target_names)
```

# Applying Train test Split 

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

# Training The MOdel WIthout Hyper Parameter Tuning 

```python
dr=DecisionTreeRegressor(random_state=32)
```

```python
dr.fit(x_train,y_train)
```

```python
y_pred=dr.predict(x_test)
print('R2 Score=',r2_score(y_test,y_pred))
```

# Training Model WIth HyperParameter Tuning 

```python
model=DecisionTreeRegressor()
```

```python
parameters={
    'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter':['best', 'random'],
    'max_depth':[5,6,7,8,4,10],
    'min_samples_split':[50,100,20,25,23,10,30,35,45], 
    'min_samples_leaf' :[20,25,23,10,30,35,45,50], 
    'max_features':['log2','sqrt',3,4,6,7,8,5], 
    'max_leaf_nodes':[5,10,8,15,14,20,12,11,5,6,7], 
    'ccp_alpha':[0.01,.002,.0001,0.0030,0.04,0.5,.6,.7,.8]
}
```

```python
cv=GridSearchCV(model,param_grid=parameters,n_jobs=-1,cv=5,scoring='r2',return_train_score=True)
```
