```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import  accuracy_score
from sklearn.tree import plot_tree
from sklearn.preprocessing import OrdinalEncoder
```

```python
def perform(model,p,q,r,s):
    model.fit(p,q) 
    # plot_tree(model)
    y_pred=model.predict(r) 
    print(f'Accuracy Score of {model} is=\t',accuracy_score(y_pred,s))
    
```

```python
df=pd.read_csv('HR.csv')
df=df.drop(columns=['Department'])
```

# Taking Random row and cheking the  result 

```python
self_test=df.iloc[13001:,:].sample(200)
self_test
```

```python
x=self_test.drop(columns=['left'])
y=self_test['left']
```

```python
y_pred1=clf1.predict(x)
y_pred2=clf2.predict(x)
y_pred3=clf3.predict(x)
print(clf1.predict(x))
print(clf2.predict(x))
print(clf3.predict(x))
print('---------Actual---------')
print(y)
y_dataframe=pd.DataFrame({
    'RF1': y_pred1,
    'RF2': y_pred2,
    'RF3': y_pred3,
    'Actual Y': y
})
```

```python
y_dataframe=y_dataframe.reset_index(drop=True)

y_dataframe
```

```python
rf1_wrong = []
rf2_wrong = []
rf3_wrong = []
actual_y = []
count=0

def wrong(): 
    global count;
    for i in range(y_dataframe.shape[0]): 
        if ((y_dataframe['RF1'][i] != y_dataframe['RF2'][i]) or 
        (y_dataframe['RF2'][i] != y_dataframe['RF3'][i]) or 
        (y_dataframe['RF1'][i] != y_dataframe['RF3'][i])):
            print(i)
            count=count+1
            rf1_wrong.append(y_dataframe['RF1'][i])
            rf2_wrong.append(y_dataframe['RF2'][i])
            rf3_wrong.append(y_dataframe['RF3'][i])
            actual_y.append(y_dataframe['Actual Y'][i])    
wrong()
wrong_df=pd.DataFrame({
            'RF1-Wrong':rf1_wrong,
            'RF2-Wrong':rf2_wrong,
            'RF3-Wrong':rf3_wrong,
            'Actual-Y':actual_y}).reset_index(drop=True)

print(wrong_df)
print(f'\nTotal Different Predictions={count}')
```

```python
#  Accuracy On Testing  0.905,0.935,0.875

print('accuracy Scores are',accuracy_score(y,y_pred1),accuracy_score(y,y_pred2),accuracy_score(y,y_pred3),end='\t')
```
