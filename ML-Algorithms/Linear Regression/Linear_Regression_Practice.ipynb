```python
from sklearn.model_selection import train_test_split


x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.9,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()

model.fit(x_train,y_train)
```

```python
model.predict([[6.63]])
```

```python
def input_function(): 
    num=int(input('Enter the number of cgpa='))
    gpa=[]
    for i in range(num):
        cgpa=float(input('Enter the value of cgpa to predict package='))
        gpa.append(cgpa)

    gpa=np.array(gpa).reshape(-1,1)
    gpa=pd.DataFrame(gpa, columns=["cgpa"])
    predictions = model.predict(gpa)
    predictions=pd.DataFrame(predictions,columns=['package'])
    df_answer=pd.concat([gpa,predictions],axis=1)
    print(df_answer)
```

```python
input_function()
```

```python
sns.scatterplot(x='cgpa',y='package',data=df)
plt.plot(x_train,model.predict(x_train),color='red') 
plt.savefig()
plt.show()

```
