```python

class MBGcalculator:
    def __init__(self, learning_rate=0.01, epochs=100, mini_batch=32):
        self.lr = learning_rate
        self.ep = epochs
        self.mb = mini_batch
        self.coeff_ = None
        self.intercept = None  # Retained your variable name

    def fit(self, x, y):
        # Initialize coefficients to match feature count
        self.intercept = 0  # Original name "intercept" preserved
        self.coeff_ = np.ones(x.shape[1])  # Critical fix: x.shape[1] not x.shape[0]

        for i in range(self.ep):
            # Shuffle data at every epoch (fixes ordering bias)
            indices = np.random.permutation(x.shape[0])  # Corrected typo
            x_shuffled = x[indices]
            y_shuffled = y[indices]

            if self.mb == 1:
                # Stochastic Gradient Descent (SGD)
                for j in range(x_shuffled.shape[0]):
                    x_j = x_shuffled[j]
                    y_j = y_shuffled[j]

                    # Prediction and gradients (retained your formula style)
                    y_pred = self.intercept + np.dot(x_j, self.coeff_)
                    inter_gradient = -2 * (y_j - y_pred)  # Original variable name
                    coeff_gradient = -2 * (y_j - y_pred) * x_j  # Fixed element-wise multiplication

                    # Parameter updates (no duplicate updates)
                    self.intercept -= self.lr * inter_gradient
                    self.coeff_ -= self.lr * coeff_gradient

            else:
                # Mini-Batch Gradient Descent (fixed batch logic)
                num_batches = x_shuffled.shape[0] // self.mb
                for batch in range(num_batches):
                    # Get mini-batch (fixed indexing)
                    start = batch * self.mb
                    end = start + self.mb
                    x_batch = x_shuffled[start:end]
                    y_batch = y_shuffled[start:end]

                    # Batch prediction and gradients
                    y_pred = self.intercept + np.dot(x_batch, self.coeff_)
                    error = y_batch - y_pred

                    # Gradients averaged over batch (retained your naming)
                    inter_gradient = -2 * np.mean(error)
                    coeff_gradient = -2 * np.dot(error, x_batch) / x_batch.shape[0]

                    # Update parameters
                    self.intercept -= self.lr * inter_gradient
                    self.coeff_ -= self.lr * coeff_gradient

        print(f"Final coefficients: {self.coeff_}, Intercept: {self.intercept}")

    def predict(self, x):
        # Corrected prediction formula (critical fix)
        return np.dot(x, self.coeff_) + self.intercept  # NOT element-wise multiplication
```
