```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt


```

```python
df=pd.read_csv('salary.csv')
df 
```

```python
#Training model for linear regression 

from sklearn.linear_model import LinearRegression

ml=LinearRegression()

x=df[['YearsExperience']]
y=df['Salary']
x_trained=ml.fit(x,y,sample_weight=None)
x_trained

```

```python
#Slope
x_trained.coef_

```

```python
#Bias Term / Intercept 
x_trained.intercept_
```

```python
plt.plot(x,ml.predict(x),color='Red')
plt.savefig('ml_predicted_value.png')
plt.show()
```

# Calculating Gradient Descent

# Lets initialize to check how the gradient Descent works  

```python
# Suppose the m is constant but we have to calculate the value of intercept now 
# Suppose c=100
b=100
m=9450
```

```python
y_pred=((m*x.to_numpy().ravel())+b)
y_pred
```

```python
plt.plot(x,ml.predict(x),color='Red')
plt.plot(x,y_pred,color='Black')
plt.savefig('initialized_value.png')
plt.show()
```

```python
# SLope 
slope=-2*(np.sum(y-y_pred))/df.shape[0]
slope
```

```python
# Lets consider learning Rate to be 0.1 
lr=0.1
step_size=lr*slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred1=((m*x.to_numpy().ravel())+b) 
y_pred1
```

```python
plt.plot(x,ml.predict(x),color='Red')
plt.plot(x,y_pred,color='Black')
plt.plot(x,y_pred1,color='Blue')
plt.savefig('iteration1.png')
plt.show()
```
