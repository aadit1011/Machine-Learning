```python
class StochasticGradientDescent: 
    def __init__(self,learning_rate=0.01,epochs=100):
        self.lr=learning_rate
        self.ep=epochs
        self.coeff_=None 
        self.intercept=None 
        
    def fit(self,x,y): 
        self.intercept=0 
        self.coeff_=np.zeros(x.shape[1])
        
        for i in range(self.ep):
           indices = np.arange(x.shape[0])  # Create a list of indices
           np.random.shuffle(indices)  # Shuffle indices at the start of each epoch

           for j in indices:  # Iterate through shuffled indices
                y_pred = self.intercept + np.dot(x[j], self.coeff_)
                inter_gradient = -2 * (y[j] - y_pred)
                self.intercept = self.intercept - (self.lr * inter_gradient)
                coeff_gradient = -2 * np.dot((y[j] - y_pred), x[j])
                self.coeff_ = self.coeff_ - (self.lr * coeff_gradient)
                self.coeff_=self.coeff_-(self.lr*coeff_gradient)
                # print(f'\n\n\nThe variables after {j+1}th row calculation are== \n',self.coeff_,self.intercept,len(self.coeff_),end='\n')
            # print(f'\n\n\nAfter the {i}th epoch the value of th e coefficients are===\n',self.coeff_,self.intercept,len(self.coeff_),end='\n')
        print(f'The final value of the variables are==\n\n\n',self.coeff_,self.intercept,len(self.coeff_),end='\n')

    def predict(self,x):
        return np.dot(x,self.coeff_)+self.intercept

    
```

```python
sgd=StochasticGradientDescent(0.01,30)
x_trained_sgd=sgd.fit(x_train,y_train) 
```

```python
y_pred=sgd.predict(x_test)
```
