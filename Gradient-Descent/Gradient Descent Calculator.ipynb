# Varying Both m and b 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

class GradientDescentCalculator:
    def __init__(self, learning_rate, epochs):
        self.m = 0  # Initialize slope
        self.b = 100  # Initialize intercept
        self.lr = learning_rate
        self.ep = epochs

    def calculator(self, x, y):
        x = x.to_numpy().ravel()
        n = len(y)
        for _ in range(self.ep):
            y_pred = self.m * x + self.b
            # Gradients
            gradient_m = (-2/n) * np.sum(x * (y - y_pred))
            gradient_b = (-2/n) * np.sum(y - y_pred)
            self.m -= self.lr * gradient_m
            self.b -= self.lr * gradient_b
            print(gradient_b,gradient_m,self.b,self.m)
        print(self.b,self.m)
            
```

```python
# Initialize calculator with adjusted learning rate (larger for visualization)
gd = GradientDescentCalculator(learning_rate=0.01, epochs=4000)
gd.calculator(x, y)  
```
